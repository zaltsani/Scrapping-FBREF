```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

# Opponent Shooting Stats Premier League 2024-25

```python
url = "https://fbref.com/en/comps/9/Premier-League-Stats"
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, "html.parser")
table_shot_against = soup.find(id="stats_squads_shooting_against")
table_body = table_shot_against.find("tbody")
rows = table_body.find_all("tr")
```

```python
table_head = table_shot_against.find("thead")
head = table_head.find_all("tr")[1]
rows_head = head.find_all("th")
columns = []
for row in rows_head:
    name = row.get_text()
    columns.append(name)
df = pd.DataFrame(columns=columns)
```

```python
column_name = []
for row in rows_head:
    name = row.get_text()
    column_name.append(name)
```

```python
for row in rows:
    team_name = row.find('a').get_text().split(' ')[1]
    cols = row.find_all('td')
    data = [col.get_text() for col in cols]
    data.insert(0, team_name)
    df_length = len(df)
    df.loc[df_length] = data
df
```
